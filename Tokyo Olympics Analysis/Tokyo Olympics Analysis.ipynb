```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
medals = pd.read_csv('Medals.csv')
medals
```

```python
medals.dropna(inplace=True)
```

## Top 30 countries by total medal count

```python
#m_total = medals.Total.sort_values(ascending=False)
#m_total = m_total.iloc[:30]
m_total = medals.sort_values(by=['Total'], ascending=False)
m_total = m_total.iloc[:30]
m_team = m_total.Team
m_total1 = m_total.Total
```

```python
cmap = plt.cm.tab10
colors = cmap(np.arange(len(medals)) % cmap.N)
plt.figure(figsize=(15,6))
plt.xticks(rotation='vertical', size='8')
plt.bar(m_team, m_total1, color=colors)
plt.xlabel('Countries')
plt.ylabel('Medals')
plt.show()
```

## Total Share in Total Medals by Top10 Countries

```python
m_total1 = m_total1.loc[:10]
m_c = m_total.Team
```

```python
m_total1.plot(kind='pie', labels = m_c, autopct='%0.1f%%', radius=2)
plt.show()
```

## Top 20 Countries with Gold, Silver, Bronze

```python
#Gold Medals
gold = medals.sort_values(by=['Gold'],ascending=False)
colors = cmap(np.arange(len(medals)) % cmap.N)
gold = gold.iloc[:20]
plt.bar(gold.Team, gold.Gold, color=colors)
plt.xticks(rotation='vertical', size=10)
plt.title('Gold Medals Share')
plt.xlabel('Medals')
plt.ylabel('Countries')
plt.show()

#Silver Medals
silver = medals.sort_values(by=['Silver'],ascending=False)
colors = cmap(np.arange(len(medals)) % cmap.N)
silver = silver.iloc[:20]
plt.bar(silver.Team, silver.Silver, color=colors)
plt.xticks(rotation='vertical', size=10)
plt.title('Silver Medals Share')
plt.xlabel('Medals')
plt.ylabel('Countries')
plt.show()

#Bronze Medals
bronze = medals.sort_values(by=['Bronze'],ascending=False)
colors = cmap(np.arange(len(medals)) % cmap.N)
bronze = bronze.iloc[:20]
plt.bar(bronze.Team, bronze.Bronze, color=colors)
plt.xticks(rotation='vertical', size=10)
plt.title('Bronze Medals Share')
plt.xlabel('Medals')
plt.ylabel('Countries')
plt.show()
```
